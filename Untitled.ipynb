```python
import random
# Function to load questions from a file
def load_questions(file_name):
    questions = []
    with open(file_name, 'r') as file:
        lines = file.readlines()
        question = None
        for line in lines:
            line = line.strip()
            if line.startswith("Q: "):
                if question:
                    questions.append(question)
                question = {"question": line[3:], "options": []}
            elif line.startswith("A: "):
                question["answer"] = int(line[3:])  # Assuming answer is an integer
            elif line.startswith("O: "):
                question["options"].append(line[3:])
        if question:
            questions.append(question)
    return questions

# Function to generate a random quiz
def generate_quiz(questions, num_questions):
    if num_questions > len(questions):
        print("Warning: Insufficient questions available. Cannot generate the requested quiz.")
        return []
    return random.sample(questions, num_questions)


# Function to grade a quiz
def grade_quiz(quiz, key):
    score = 0
    for i, question in enumerate(quiz):
        print(f"Q{i + 1}: {question['question']}")
        for j, option in enumerate(question["options"]):
            print(f"{j + 1}. {option}")
        user_answer = int(input("Your answer (enter the option number): "))
        if user_answer == question["answer"]:
            score += 1
    return score

if __name__ == "__main__":
    question_file = "question.txt"  # Replace with your question file
    num_questions = 5  # Number of questions in the quiz
    questions = load_questions(question_file)

    while True:
        print("Quiz Maker")
        input("Press Enter to start a quiz...")
        
        # Generate a random quiz
        quiz = generate_quiz(questions, num_questions)
        
        # Grade the quiz
        score = grade_quiz(quiz, questions)
        print(f"Your score: {score}/{num_questions}\n")

        play_again = input("Do you want to play again? (yes/no): ").lower()
        if play_again != "yes":
            break
```
